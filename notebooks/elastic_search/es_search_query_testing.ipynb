```python
def get_results_list(results):
    results = results['hits']['hits']

    result_list = []
    for doc_data in results:
        doc_dict = dict()

        doc_dict['text'] = doc_data['_source']['autocomplete']

        result_list.append(doc_dict)
        
    return result_list
```

```python
query_body = {
      "fuzzy": {
          "contents.de": "rader"
      }
}

query_body_autocomplete = {
    "match": {
      "autocomplete": {
        "query": "Rad",
      }
  }
}

results = es.search(index=index, query=query_body_autocomplete, size=10)
get_results_list(results)
```
