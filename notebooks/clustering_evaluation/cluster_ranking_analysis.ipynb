```python
import pandas as pd
import os
import string
import numpy as np
import json

import matplotlib.pyplot as plt
import seaborn as sns 
%matplotlib inline

import warnings
warnings.filterwarnings('ignore')
```

```python
filepath = os.getcwd()+'/../dataframes/ranking_data/'
filename_list_1 = ['mixed_reality_IR2_ranking.pkl', 'mixed_reality_IR3_ranking.pkl', 'mixed_reality_IR4_ranking.pkl']
filename_list_2 = ['visualisierung_IR2_ranking.pkl', 'visualisierung_IR3_ranking.pkl', 'visualisierung_IR4_ranking.pkl']
```

```python
def get_ranking_comparision(filename_list, col_name):
    
    mr_2 = pd.read_pickle(filepath+filename_list[0])
    mr_3 = pd.read_pickle(filepath+filename_list[1])
    mr_4 = pd.read_pickle(filepath+filename_list[2])
    
    mr_2_topic_list = mr_2[col_name].tolist()
    mr_3_topic_list = mr_3[col_name].tolist()
    mr_4_topic_list = mr_4[col_name].tolist()
    
    topic_df = pd.DataFrame([mr_2_topic_list, mr_3_topic_list, mr_4_topic_list])
    topic_df = topic_df.T
    topic_df.columns = ['query ranking', 'template', 'document count']
    
    return topic_df
```
